```python
def get_garden_plan(insect_name, sun=None, growth=None):
    result = df[df["Insect"] == insect_name]
    
    if sun:
        result = result[result["Sun"] == sun]
        
    if growth:
        result = result[result["Growth"] == growth]
    
    if result.empty:
        print("No data found. Check spelling or filter choice.")
        return
    
    host = result[result["Stage"] == "Host"][["Plant", "Botanical Name"]].drop_duplicates()
    nectar = result[result["Stage"] == "Butterfly"][["Plant", "Botanical Name"]].drop_duplicates()
    
    print(f"\n🌱 Garden plan for {insect_name}")
    
    if sun:
        print(f"☀️ Sun requirement filter: {sun}")
    if growth:
        print(f"🌿 Growth filter: {growth}")
    
    print("\n🐛 Host plants (must-have):")
    for _, row in host.iterrows():
        print(f"- {row['Plant']} ({row['Botanical Name']})")
    
    print("\n🦋 Nectar plants (support adults):")
    for _, row in nectar.iterrows():
        print(f"- {row['Plant']} ({row['Botanical Name']})")
    
    print("\n✨ Suggested starter planting mix:")
    
    host_mix = host.head(2)
    nectar_mix = nectar.head(3)
    
    print("Start with:")
    for _, row in host_mix.iterrows():
        print(f"- 1 host plant: {row['Plant']} ({row['Botanical Name']})")
    
    for _, row in nectar_mix.iterrows():
        print(f"- 1 nectar plant: {row['Plant']} ({row['Botanical Name']})")
```

```python
def get_garden_plan(insect_name, sun=None, growth=None, mode="starter"):
    result = df[df["Insect"] == insect_name]
    
    if sun:
        result = result[result["Sun"] == sun]
        
    if growth:
        result = result[result["Growth"] == growth]
    
    if result.empty:
        print("No data found. Check spelling or filter choice.")
        return
    
    host = result[result["Stage"] == "Host"][["Plant", "Botanical Name"]].drop_duplicates()
    nectar = result[result["Stage"] == "Butterfly"][["Plant", "Botanical Name"]].drop_duplicates()
    
    print(f"\n🌱 Garden plan for {insect_name}")
    
    if sun:
        print(f"☀️ Sun requirement filter: {sun}")
    if growth:
        print(f"🌿 Growth filter: {growth}")
    
    if mode == "full":
        print("\n🐛 Host plants (must-have):")
        for _, row in host.iterrows():
            print(f"- {row['Plant']} ({row['Botanical Name']})")
        
        print("\n🦋 Nectar plants (support adults):")
        for _, row in nectar.iterrows():
            print(f"- {row['Plant']} ({row['Botanical Name']})")
    
    elif mode == "starter":
        print("\n✨ Suggested starter planting mix:")
        
        host_mix = host.head(2)
        nectar_mix = nectar.head(3)
        
        print("Start with:")
        for _, row in host_mix.iterrows():
            print(f"- 1 host plant: {row['Plant']} ({row['Botanical Name']})")
        
        for _, row in nectar_mix.iterrows():
            print(f"- 1 nectar plant: {row['Plant']} ({row['Botanical Name']})")
    
    else:
        print("Mode not recognized. Use 'full' or 'starter'.")
```

```python
get_garden_plan("Monarch", sun="Full", growth="Herbaceous", mode="starter")
```
